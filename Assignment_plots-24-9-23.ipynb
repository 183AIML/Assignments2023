```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("D:\Datasets\Bangalore  house data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['location'].fillna('sv nagar',inplace=True)
```

```python
df.isnull().sum()
```

```python
df['society'].fillna('no society',inplace=True)
```

```python
df.isnull().sum()
```

```python
df['size'].fillna('2bhk',inplace=True)
```

```python
df.isnull().sum()
```

```python
df['bath'].fillna('2.0',inplace=True)
```

```python
df.isnull().sum()
```

```python
df['balcony'].fillna(1.0,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
plt.figure(figsize=(30,15))#hist
sns.histplot(x='size',data=df)
plt.show()
```

```python
plt.figure(figsize=(30,25))#barplot
sns.barplot(x='size',y='price',data=df)
plt.show()
```

```python
plt.figure(figsize=(15,30))#boxplot
sns.boxplot(y='price',data=df)
plt.show()
```

```python
plt.figure(figsize=(8,10))#boxplot
sns.boxplot(x='price',y='size',data=df)
plt.show()
```

```python
plt.figure(figsize=(8,10))
sns.scatterplot(x='bath',y='price',data=df)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.lineplot(x='price',y='size',data=df)#line
plt.show()
```
