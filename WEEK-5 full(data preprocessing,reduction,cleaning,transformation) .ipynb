# TO READ THE FILE

```python
ht_wt_data=pd.read_csv("C:/Documents/500_Person_Gender_Height_Weight_Index.csv")
ht_wt_data.head(5)
```

# DROPING THE COLOUMN

```python
ht_wt_data.drop('Index',inplace=True,axis=1)
print(ht_wt_data.head(5))
print(ht_wt_data.shape)
```

# MAX AND MIN OF THE HEIGHT AND WEIGHT

```python
ht_max=ht_wt_data['Height'].max()
print('max_height_is=',ht_max)
wt_max=ht_wt_data['Weight'].max()
print('max_weight_is=',wt_max)
ht_min=ht_wt_data['Height'].min()
print('min_height_is=',ht_min)
wt_min=ht_wt_data['Weight'].min()
print('min_weight_is=',wt_min)
```

# MEAN AND MODE OF THE HEIGHT AND WEIGHT

```python
ht_mean=ht_wt_data['Height'].mean()
print('mean_height_is=',ht_mean)
wt_mean=ht_wt_data['Weight'].mean()
print('mean_weightis=',wt_mean)
ht_mode=ht_wt_data['Height'].mode()
print('mode_height_is=',ht_mode)
wt_mode=ht_wt_data['Weight'].mode()
print('mode_weight_is=',wt_mode)
```

# IT TAKES THE PARTICULAR COLUMN OF 5 ROWS

```python
height=ht_wt_data['Height']
type(height)
height.head(5)
```

# SORTING TO GET THE MEADIAN OF THE HEIGHT

```python
height_median=height.sort_values().reset_index(drop=True)
print(height_median.median())
```

# MEADIAN OF THE HEIGHT

```python
height1=height_median.median()
print(height1)
```

# IT TAKES THE PARTICULAR COLUMN OF 5 ROWS

```python
weight=ht_wt_data['Weight']
type(weight)
weight.head(5)
```

# SORTING TO GET THE MEADIAN OF THE WEIGHT

```python
weight =weight.sort_values().reset_index(drop=True)
print(weight.median())
```

# MEADIAN OF THE HEIGHT

```python
weight1=weight.median()
print(weight1)
```

# HISTOGRAM OF THE MEAN,MEDIAN OF WEIGHT AND HEIGHT

```python
plt.figure(figsize =(12,8))
height.hist(bins=30)
plt.axvline(ht_mean,label='mean',color='g')
plt.axvline(height1,label='median',color='r')
plt.legend()
plt.show()
```

```python
plt.figure(figsize =(12,8))
height.hist(bins=30)
plt.axvline(wt_mean,label='mean',color='g')
plt.axvline(weight1,label='median',color='r')
plt.legend()
plt.show()
```
